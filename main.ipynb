```python
import pandas as pd
import numpy as np
```

<h1 align='center'>Importing Data</h1>

```python
df = pd.read_csv("capstone.csv")
df.head()
```

<h1 align='center'>Preprocessing</h1>

```python
df.isnull().sum()
```

```python
result1 = df[df['Levy'] == '-']
result1.shape
```

```python
df.replace('-', np.nan,inplace=True)
```

```python
df.fillna(0, inplace=True)
df.head(10)
```

```python
def split(X):
    X=str(X)
    X=X.split(" ")[0]
    return float(X)
df['Engine volume']=df['Engine volume'].apply(lambda X: split(X)) 
df['Mileage']=df['Mileage'].apply(lambda X: split(X)) 
df['Mileage'].astype('int64')
df.head()
```

```python
df['Leather interior'] = df['Leather interior'].apply(lambda x: 0 if x == 'No' else 1)
df.head()
```

```python
dummy = pd.get_dummies(df[['Manufacturer','Model','Category','Fuel type','Gear box type','Drive wheels','Doors','Wheel','Color']],drop_first=True,dtype=int)
```

```python
dummy.shape
```

```python
df= pd.concat([df, dummy],axis=1)
df.head()
```

```python
df.drop(['Manufacturer','Category','Fuel type','Gear box type','Drive wheels','Doors','Wheel','Color','ID'],axis=1,inplace=True)
```

```python
df.drop(['Model'],axis=1,inplace=True)
df.head()
```

```python
df.to_csv('mydata.csv')
```

<h1 align='center'>Normalize the data</h1>

```python
from sklearn.preprocessing import normalize
```

```python
columns = df.columns
df_n = normalize(df)
df_n
```

```python
scaled_df = pd.DataFrame(df_n, columns=columns)
scaled_df.head()
```

```python
df_n=scaled_df[['Price','Levy','Prod. year','Leather interior','Engine volume','Mileage','Cylinders','Airbags']]
df_n.head()
```

```python
df.drop(['Price','Levy','Prod. year','Leather interior','Engine volume','Mileage','Cylinders','Airbags'],axis=1,inplace=True)
```

```python
df_final = pd.concat([df_n,df],axis=1)
df_final.head()
```

<h1 align='center'>Training</h1>

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df_final
y=df_final['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

<h1 align='center'>Hyper Parameter Tuning</h1>

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
model_params = {
    'Linear Regression': {
        'model': LinearRegression(),
        'params' : {}  
    },
    'Lasso': {
        'model': Lasso(),
        'params' : {
            'alpha': [1,5,10]
        }
    },
    'Ridge' : {
        'model': Ridge(),
        'params': {
            'alpha': [1,5,10]
        }
    },
    'SVR' : {
        'model': SVR(),
        'params': {
            'kernel':['linear','poly','rbf','sigmoid'],
            'C': [1,5,10]
        }
    },
    'DecisionTreeRegressor' : {
        'model': DecisionTreeRegressor(),
        'params': {
            'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
        }
    },
    'KNeighborsRegressor' : {
        'model': KNeighborsRegressor(),
        'params': {
            'n_neighbors': [3,5,7,10]
        }
    },
    'RandomForestRegressor' : {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [50,100,150]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
scores=list()
for model,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5)
    clf.fit(x,y)
    scores.append({
        'model':model,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
    
```

```python
df_scores=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df_scores.sort_values(by=['best_score'],ascending=False)
```

```python
import matplotlib.pyplot as plt

y_pred = model.predict(x_test)

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue',label='Actual vs. Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, color='red', label='Ideal Prediction')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs. Predicted Prices')
plt.legend()
plt.grid(True)
plt.show()
```
